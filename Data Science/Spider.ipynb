# Spider is something that crawls over the web and collects data

```python
import requests                         # collects the webpage data object
from bs4 import BeautifulSoup           # class that converts object into readable object
```

```python
addr = 'https://en.wikipedia.org/wiki/September_10'        # webpage address
page = requests.get(addr)                               # collects data of webpage
print(page.status_code)                                 # status code 200 means success
```

```python
soup = BeautifulSoup(page.text)     # soup => object of class BeautifulSoup
```

```python
soup                # display all content of webpage
```

```python
soup.title          # display title of webpage
```

```python
soup.find('div',attrs={'id':'mp-otd'})  

''' for fetching one div we use find
    for fetching multiple div we use findall
'''
```

# Use data to get only text

```python
data = soup.find('div',attrs={'id':'mp-otd'}).text
print(data)
```

# Lets fetch data from another Wiki Page
# September 10

```python
datalist = soup.find_all('ul')
for item in datalist:
    print(item.text)
```
