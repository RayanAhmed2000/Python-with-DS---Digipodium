- selecting rows and columns
    - rows
        - `loc` - gets us the data using named indexes()
        - `iloc` - gets us the data using numeric indexes
    - column (dictionary style)
        - `df[single]` - requires name of column
        - `df[multiple]` - requires a list of columns (`multiple` is a list of named cols)

### single row


```python
weather.iloc[0]  #first row
```

```python
weather.iloc[-1]  #last row
```

### 25 - 50 records

```python
weather.iloc[25:51]
```
